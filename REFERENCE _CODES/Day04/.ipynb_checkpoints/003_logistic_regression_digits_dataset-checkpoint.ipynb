# Logistic Regression Demo with scikit-learn and Digits Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model, metrics
from sklearn.model_selection import train_test_split
import seaborn as sns
```

## Load the Digits Dataset

The digits dataset consists of 8x8 images of handwritten digits.

```python
# Load the digits dataset
digits = datasets.load_digits()
X = digits.data
y = digits.target
```

```python
X.shape, y.shape
```

## Visualize Some Sample Digits

```python
# Plot the first 10 images from the dataset
fig, axes = plt.subplots(1, 10, figsize=(12, 2))
for i, ax in enumerate(axes):
    ax.imshow(X[i].reshape(8, 8), cmap='gray')
    ax.set_title(f"Label: {y[i]}")
    ax.axis('off')
plt.show()
```

## Split the Data into Training and Test Sets

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.4, random_state=42, stratify=y
)
print(f"Training samples: {X_train.shape[0]}, Test samples: {X_test.shape[0]}")
```

## Train a Logistic Regression Model

```python
# Create and train the logistic regression model
reg = linear_model.LogisticRegression(max_iter=1000)
reg.fit(X_train, y_train)
```

## Evaluate the Model

```python
# Predict on the test set
y_pred = reg.predict(X_test)

# Calculate accuracy
accuracy = metrics.accuracy_score(y_test, y_pred)
print(f"Test set accuracy: {accuracy:.2%}")
```

## Confusion Matrix Visualization

```python
# Plot confusion matrix
cm = metrics.confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", 
            xticklabels=digits.target_names, yticklabels=digits.target_names)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()
```

## Visualize Some Predictions

```python
# Show 10 random test images with predictions
indices = np.random.choice(len(X_test), 10, replace=False)
fig, axes = plt.subplots(1, 10, figsize=(12, 2))
for i, idx in enumerate(indices):
    ax = axes[i]
    ax.imshow(X_test[idx].reshape(8, 8), cmap='gray')
    ax.set_title(f"P:{y_pred[idx]}\nT:{y_test[idx]}")
    ax.axis('off')
plt.show()
```
